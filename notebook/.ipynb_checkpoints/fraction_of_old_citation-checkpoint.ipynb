```python
import os
import ijson
import json
import math
import statistics
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
%matplotlib inline
```

```python
non_cs_cats = [
    'math.AP',
    'hep-ph',
    'econ.GN',  # (General Economics) Economics
    'eess.SP',  # (Signal Processing) Electrical Engineering and Systems Science
    'q-bio.PE',  # (Populations and Evolution) Quantitative Biology
    'q-fin.ST',  # (Statistical Finance) Quantitative Finance
    'stat.ME' # (Methodology) Statistics
 ]

cs_non_ai_cats =[
"cs.CR",# (Cryptography and Security)
"cs.IT",# (Information Theory)
"cs.NI",# (Networking and Internet Architecture)
"cs.DS",# (Data Structures and Algorithms)
]

cs_ai =[
"cs.CL",# (Cryptography and Security)
"cs.AI",# (Information Theory)
"cs.LG",# (Networking and Internet Architecture)
"cs.CV",# (Data Structures and Algorithms)
]

x = [i for i in range(2013, 2023)]
```

### Percentage of old citations

```python
# fraction of old citation over the year over. Individual and average over all 

cats = non_cs_cats + cs_non_ai_cats + cs_ai

cat_num = {}
for cat in cats:
    cat_num[cat] = []
    year_percentage = []
    for year in range(2013, 2023):
        print(cat, year)
        per_old_ref_all_year = []
        if os.path.exists(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json")):
            with open(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json"), "r") as fp:
                content = json.load(fp)
                per_old_ref_all = []
                for paper in content:
                    num_old_ref = 0
                    if paper["reference"]:
                        for ref in paper["reference"]:
                            if ref["year"]:
                               if ref["year"] <= (year - 10):
                                   num_old_ref +=1
                            per_old_ref = num_old_ref/len(paper["reference"])
                            per_old_ref_all.append(per_old_ref)
            if per_old_ref_all:
                per_old_ref_all_year.append(statistics.mean(per_old_ref_all)*100)
            else:
                per_old_ref_all_year.append(0)
        cat_num[cat].append(per_old_ref_all_year)
```

```python
with open("percentage_old_paper_clean.json", "w") as fp:
    json.dump(cat_num, fp)
```

```python
with open("percentage_old_paper_clean.json", "r") as fp:
    cat_num = json.load(fp)
```

```python
# Subplots 
x = [i for i in range(2013, 2023)]
fig, (ax1, ax2, ax3) = plt.subplots(3, sharex=True,sharey=True,figsize=(10, 9))
fig.supxlabel('Year', fontsize=20)
fig.supylabel('Percentage of old papers',fontsize=20)

for cat, num in cat_num.items():
    if cat in non_cs_cats:
        ax1.plot(x, num, '--', label=cat,linewidth=2, marker='o')
        ax1.legend(ncol=4, prop={'size': 11})
        ax1.yaxis.set_tick_params(labelsize=16)
        ax1.set_title("non_cs",fontsize=16)
    elif cat in cs_non_ai_cats:
        ax2.plot(x, num, ':', label=cat,linewidth=2, marker='o')
        ax2.legend(ncol=4, prop={'size': 11})
        ax2.yaxis.set_tick_params(labelsize=16)
        ax2.set_title("cs_non_ai",fontsize=16)
    elif cat in cs_ai:
        ax3.plot(x, num, label=cat,linewidth=2, marker='o')
        ax3.legend(ncol=4, prop={'size': 11})
        ax3.yaxis.set_tick_params(labelsize=16)
        ax3.set_title("cs_ai",fontsize=16)
        
plt.xticks(np.arange(min(x), max(x)+1, 1.0), rotation=35,fontsize=16)
ax1.yaxis.set_major_formatter(mtick.PercentFormatter())
#plt.xlabel("Year")
#plt.ylabel("Percentage of old papers")
#plt.legend(ncol=3, prop={'size': 6})
plt.savefig("percentage_old_paper.svg", format="svg")
plt.show()
```

```python
# Age of citation
```

```python
def normalize_age(ref_years, year, threshold=0):
    ages = []

    for pub_year in ref_years:
        if pub_year and math.isnan(float(pub_year)) is not True:
            if int(pub_year) <= year : # remove errors
                if year - int(pub_year) > threshold:
                    pub_year = year-threshold
                
                elif threshold == 0 and int(pub_year) <= year: 
                    ages.append(year-int(pub_year))
    return ages
```

```python
def get_age(ref_years, year):
    ages = []
    for pub_year in ref_years:
        if pub_year and math.isnan(float(pub_year)) is not True:
            if int(pub_year) <= year : # remove errors
                ages.append(year-int(pub_year))
    return ages
```

### Mean citation age

```python
cats = non_cs_cats + cs_non_ai_cats + cs_ai

cat_mean = {}
for cat in ["cs.AI"]:
    cat_mean[cat] = []
    for year in range(2013, 2023):
        print(cat, year)
        year_mean = []
        if os.path.exists(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json")):
            count = 0
            with open(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json"), "r") as fp:
                content = json.load(fp)
                for paper in content:
                    references = []
                    if paper["reference"]:
                        for ref in paper["reference"]:
                            if ref["year"]:
                                references.append(ref["year"])
                    if references:
                        ages = get_age(references, year)
                        if ages:
                            mean = statistics.mean(ages)
                            year_mean.append(mean)
        if year_mean:
            cat_mean[cat].append(statistics.mean(year_mean))
        else:
            cat_mean[cat].append(0)
```

### Median  citation age

```python
cats = non_cs_cats + cs_non_ai_cats + cs_ai

cat_median = {}
for cat in cats:
    cat_median[cat] = []
    for year in range(2013, 2023):
        print(cat, year)
        year_median = []
        if os.path.exists(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json")):
            with open(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json"), "r") as fp:
                content = json.load(fp)
                for paper in content:
                    references = []
                    if paper["reference"]:
                        for ref in paper["reference"]:
                            if ref["year"]:
                                references.append(ref["year"])
                    if references:
                        ages = get_age(references, year)
                        if ages:
                            median = statistics.median(ages)
                            year_median.append(median)
        if year_median:
            cat_median[cat].append(statistics.median(year_median))
        else:
            cat_median[cat].append(0)
```

### Mean AoC mean by cat

```python
with open("mean_AoC_clean.json", "r") as fp:
    cat_mean = json.load(fp)
```

```python
non_cs, cs_non_ai, cs_ai_ = [], [], []

avg_cat_mean= {}
for cat, num in cat_mean.items():
    if cat in non_cs_cats:
        non_cs.append(num)
        data = np.array(non_cs)
        non_cs_avg = np.average(data, axis=0)
        avg_cat_mean["non_cs"] = non_cs_avg
    elif cat in cs_non_ai_cats:
        cs_non_ai.append(num)
        data = np.array(cs_non_ai)
        cs_non_ai_avg = np.average(data, axis=0)
        avg_cat_mean["cs_non_ai"] = cs_non_ai_avg
    elif cat in cs_ai:
        cs_ai_.append(num)
        data = np.array(cs_ai_)
        cs_ai_avg = np.average(data, axis=0)
        avg_cat_mean["cs_ai"] = cs_ai_avg
```

```python
x = [year for year in range(2013, 2023)]
fig, ax = plt.subplots(1, sharex=True,sharey=True,figsize=(5, 5))

fig.supxlabel('Year', fontsize=12)
fig.supylabel('Mean AoC',fontsize=12)

for cat, num in avg_cat_mean.items():
    if cat == "non_cs":
        ax.plot(x, num, label="non_cs")
    elif cat == "cs_non_ai":
        ax.plot(x, num, label="cs_non_ai")
    elif cat == "cs_ai":
        ax.plot(x, num, label="cs_ai")
ax.legend()        
plt.xticks(np.arange(min(x), max(x)+1, 1.0), rotation=35,fontsize=9)
plt.savefig("mean_old_paper_by_cat.svg", format="svg")
plt.show()
```

### mean aoc of old citation in influential citation

```python
cats = non_cs_cats + cs_non_ai_cats + cs_ai

cat_mean_i = {}
for cat in cats:
    cat_mean_i[cat] = []
    for year in range(2013, 2023):
        print(cat, year)
        year_mean = []
        if os.path.exists(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json")):
            with open(os.path.join(os.path.dirname(os.getcwd()),f"data/{cat}/{year}/output_clean.json"), "r") as fp:
                content = json.load(fp)
                for paper in content:
                    references = []
                    if paper["reference"]:
                        for ref in paper["reference"]:
                            if ref["year"]:
                                if ref["isInfluential"] == True:
                                    references.append(ref["year"])
                    if references:
                        ages = get_age(references, year)
                        if ages:
                            mean = statistics.mean(ages)
                            year_mean.append(mean)
        if year_mean:
            cat_mean_i[cat].append(statistics.mean(year_mean))
        else:
            cat_mean_i[cat].append(0)
```

```python
with open("mean_AoC_infl_clean.json", "w") as fp:
    json.dump(cat_mean_i, fp)
```

```python
with open("mean_AoC_infl_clean.json", "r") as fp:
    cat_mean_i = json.load(fp)
```
