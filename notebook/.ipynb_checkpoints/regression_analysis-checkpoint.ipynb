## 1. Age of citation distribution per year per category

```python
def normalize_age(ref_years, year, threshold=50):
    ages = []

    for pub_year in ref_years:
        if pub_year and math.isnan(float(pub_year)) is not True:
            if year - int(pub_year) > threshold:
                pub_year = year-threshold
            elif int(pub_year) <= year :
                ages.append(year-int(pub_year))
    return ages
```

```python
def get_data(cat):
    threshold = 60
    
    results = []
    for year in range(2013, 2024):
        if os.path.exists(os.path.join(os.getcwd(),f"../data/{cat}/{year}/output.json")):
            with open(os.path.join(os.getcwd(),f"../data/{cat}/{year}/output.json"), "r") as fp:
                content = json.load(fp)
                reference_year = []
                for paper in content:
                    reference_year.extend([ref["year"] for ref in paper["reference"]])
    
                reference_age = normalize_age(reference_year, year, threshold)
                for age in reference_age:
                    results.append({"Year":year, "Age":age})
    df = pd.DataFrame(results)
    return df
```

```python
def plot_distribution(cat):
    df = get_data(cat)
    fig, ax = plt.subplots()
        
    fig.suptitle('Age of citations for non computer science categories')
    
    sns.violinplot(x='Year',y='Age',data=df,inner=None)
    Means = df.groupby('Year')['Age'].mean()
    Median = df.groupby('Year')['Age'].median()
    Modes = df.groupby('Year')['Age'].agg(lambda x: pd.Series.mode(x)[0])
    
    plt.setp(ax.collections, alpha=.3)
    plt.scatter(x=range(len(Means)),y=Means,label="mean")
    plt.scatter(x=range(len(Median)),y=Median,label="median")
    plt.scatter(x=range(len(Modes)),y=Modes,label="mode")
    
    ax.set_title(f'{cat}')
    ax.legend()
    axes.set_xlabel('Year')
    axes.set_ylabel('Age of citations')
    plt.show()
```

```python
def plot_all(cats):
    results = []
    for cat in cats:
        df = get_data(cat)
        mean_df = pd.DataFrame(df.groupby('Year')['Age'].mean())
        mean_df["Cat"] = [cat]*len(mean_df)
        results.append(mean_df)
            
    # Form dataframe from data
    df = pd.concat(results)
    
    # Plot multiple columns such as population and year from dataframe
    #fig = df.pivot_table(values="Age", index="Year",columns="Cat").plot(figsize=(10,5))
    fig = df.pivot_table(values="Age", index="Year",columns="Cat")
    
    return fig         
```

## 2. Calculate the coefficient of applying linear regression on categories over the years

```python
def plot_fit_regression(cat):
    fig.suptitle(f'Age of citations for {cat}')
    # Fit the regression line 
    statistics = pd.read_csv(os.path.join(os.getcwd(),f"../data/{cat}/statistics.csv"), header=0, index_col=0)
    statistics["Mean"] = statistics["Mean"].round(2)
    statistics["Mode"] = statistics["Mode"].round(2)
    statistics["Median"] = statistics["Median"].round(2)
    
    x = [i for i in range(2013, 2024)]
    y_mean = statistics["Mean"][11:22]
    y_median = statistics["Median"][11:22]
    y_mode = statistics["Mode"][11:22]

    coef_mean = np.polyfit(x,y_mean,1)
    poly1d_fn_mean = np.poly1d(coef_mean) 
    print("Coefficient for the mean of AOC", coef_mean)
    
    coef_median = np.polyfit(x,y_median,1)
    poly1d_fn_median = np.poly1d(coef_median)
    print("Coefficient for the median of AOC",coef_median)

    coef_mode = np.polyfit(x,y_mode,1)
    poly1d_fn_mode = np.poly1d(coef_mode) 
    print("Coefficient for the mode of AOC", coef_mode)
    
    plt.plot(x, y_mean, 'yo', x, poly1d_fn_mean(x), '--y', label="mean")
    plt.plot(x, y_median, 'ro', x, poly1d_fn_mode(x), '--m', label="mode")
    plt.plot(x, y_mode, 'bo', x, poly1d_fn_median(x), '--c', label="median") 
    
    plt.legend()
    plt.xticks(np.arange(2012, 2024, 1.0))
    plt.ylim(0, 25)

    plt.show()
```

## 3. Numbers of papers per year per categories

```python
all_cat = cs_ai + cs_non_ai + non_cs

results = {}
for cat in all_cat:
    num_paper = []
    for year in range(2013, 2024):
        if os.path.exists(os.path.join(os.getcwd(), f"../data/{cat}/{year}/output.json")):
            with open(os.path.join(os.getcwd(),f"../data/{cat}/{year}/output.json"), "r") as fp:
                content = json.load(fp)
                num_paper.append(len(content))
        elif os.path.exists(os.path.join(os.getcwd(),f"../data/{cat}/{year}/output_{year}_new.csv")):
            with open(os.path.join(os.getcwd(),f"../data/{cat}/{year}/output_{year}_new.csv"), "r") as fp:
                df = pd.read_csv(fp, header=0)
                num_paper.append(df.shape[0])
    results[cat] = num_paper

# create data
x = [i for i in range (2013, 2024)]

# plot lines
for cat, num_paper in results.items():
    plt.plot(x, num_paper, label = f"{cat}")
plt.legend()
plt.show()
```

## 4. Different between science parse and semantic scholar

```python
def compare(cat):
    sp_df = pd.read_csv(os.path.join(os.getcwd(), f"../data/{cat}/statistics_sp_500.csv"), header=0)
    ss_df = pd.read_csv(os.path.join(os.getcwd(), f"../data/{cat}/statistics_ss_500.csv"), header=0)
    ss_df = ss_df.loc[ss_df['Threshold'] == 20]

    results = pd.DataFrame({"Year": ss_df["Year"].tolist(),
                            "Mean_SS": ss_df["Mean"].tolist(), 
                            "Mean_SP": sp_df["Mean"].tolist(),
                            "Median_SS": ss_df["Median"].tolist(),
                            "Median_SP": sp_df["Median"].tolist()
                           })
    results.plot(x="Year", y=["Mean_SS","Mean_SP", "Median_SS", "Median_SP"], kind="bar", figsize=(4, 4), fontsize=11)
    
    results["Diff_Mean"] = np.subtract(np.array(sp_df["Mean"].tolist()),np.array(ss_df["Mean"].tolist()))
    results["Diff_Median"] = np.subtract(np.array(sp_df["Median"].tolist()),np.array(ss_df["Median"].tolist()))
    print(results)
```

```python
compare("cs.AI")
```

```python
compare("cs.CV")
```

```python
compare("econ.EM")
```

## 5. Mean, mode and median of important citation

```python
def compare_influencial(cat):
    normal_df = pd.read_csv(os.path.join(os.getcwd(), f"../data/{cat}/statistics.csv"), header=0)
    influencial_df = pd.read_csv(os.path.join(os.getcwd(), f"../data/{cat}/statistics_influencial.csv"), header=0)
    normal_df = normal_df.loc[normal_df['threshold'] == 20]

    results = pd.DataFrame({"Year": normal_df["Year"].tolist(),
                            "Mean_Nor": normal_df["Mean"].tolist(), 
                            "Mean_Infl": influencial_df["Mean"].tolist(),
                            "Median_Nor": normal_df["Median"].tolist(),
                            "Median_Infl": influencial_df["Median"].tolist()
                           })
    results.plot(x="Year", y=["Mean_Nor","Mean_Infl", "Median_Nor", "Median_Infl"], kind="bar", figsize=(4, 4), fontsize=11)
    
    results["Diff_Mean"] = np.subtract(np.array(normal_df["Mean"].tolist()),np.array(influencial_df["Mean"].tolist()))
    results["Diff_Median"] = np.subtract(np.array(normal_df["Median"].tolist()),np.array(influencial_df["Median"].tolist()))
    print(results)
```
