```python
import os
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.colors as cm
import numpy as np
```

```python
non_cs_cats = [
    #'hep-ph',
    'math.AP',
    'econ.GN',  # (General Economics) Economics
    'eess.SP',  # (General Economics) Electrical Engineering and Systems Science
    'q-bio.PE',  # (Populations and Evolution) Quantitative Biology
    'q-fin.ST',  # (Statistical Finance) Quantitative Finance
    'stat.ME' # (Methodology) Statistics
 ]

cs_non_ai_cats =[
"cs.CR",# (Cryptography and Security)
"cs.IT",# (Information Theory)
"cs.NI",# (Networking and Internet Architecture)
"cs.DS",# (Data Structures and Algorithms)
]

cs_ai =[
"cs.CL",# (Cryptography and Security)
"cs.AI",# (Information Theory)
"cs.LG",# (Networking and Internet Architecture)
"cs.CV",# (Data Structures and Algorithms)
]
```

```python
# count numbers of papers for each subcategory
cats = non_cs_cats + cs_non_ai_cats + cs_ai
cat_num = {}

for cat in cats:
    year_num = []
    for year in range(2013, 2023):
        print(cat, year)
        if os.path.exists(os.path.join(os.path.dirname(os.getcwd()), f"data/{cat}/{year}/output.json")):
            with open(os.path.join(os.path.dirname(os.getcwd()), f"data/{cat}/{year}/output.json"), "r", encoding="utf-8", errors='ignore') as fp:
                content = json.load(fp)
                year_num.append(len(content))
        elif os.path.exists(os.path.join(os.path.dirname(os.getcwd()), f"data/{cat}/{year}/output_{year}_new.csv")):
            with open(os.path.join(os.path.dirname(os.getcwd()), f"data/{cat}/{year}/output_{year}_new.csv"), "r", encoding="utf-8") as fp:
                df = pd.read_csv(fp, header=0)
                year_num.append(len(content))
    cat_num[cat] = year_num
```

```python
# create stacked bar chart for monthly temperatures
df = pd.DataFrame(cat_num)

fig, ax = plt.subplots()

colors = sns.color_palette("Paired",6)+sns.color_palette("Set2",4)+sns.color_palette("husl",4)
df.plot(kind='barh', stacked=True,color=colors, figsize=(4.5, 4.5),ax=ax)

# labels for x & y axis
plt.legend(ncol=3, prop={'size': 6})
ax.grid(True, which='major', axis='x', color='grey', zorder=0, linewidth = "0.2")
ax.set_yticklabels([i for i in range(2013, 2023)])
plt.xlabel('Paper count')
plt.ylabel('Year')

# title of plot
plt.title('Paper count by category')
```

```python
# Numbers of pairs of bars you want
N = 14

# Specify the values of blue bars (height)
blue_bar, orange_bar = [], []
for k,v in cat_num.items():
    blue_bar.append(v[0])

# Specify the values of orange bars (height)
for k,v in cat_num.items():
    orange_bar.append(v[-1])  
```

```python
# Position of bars on x-axis
ind = np.arange(N)

# Figure size
plt.figure(figsize=(4.5, 4.5))

# Width of a bar 
width = 0.3     

# Plotting
plt.bar(ind, blue_bar , width, label='2013', zorder=2)
plt.bar(ind + width, orange_bar, width, label='2022', zorder=2)

plt.xlabel('Category')
plt.ylabel('Paper count')

# xticks()
# First argument - A list of positions at which ticks should be placed
# Second argument -  A list of labels to place at the given locations
plt.xticks(np.arange(0,14, 1.0),labels=list(cat_num.keys()),rotation=90)

# Finding the best position for legends and putting it
plt.grid(True, color = "grey", linewidth = "0.2", zorder=1)
plt.legend(ncol=3, prop={'size': 6})
plt.savefig("paper_num_by_year_change.svg",format="svg")
plt.show()
```
